```julia
lines_mst = readlines("./results/tabu/mst_vs_rand/mst.txt")
    
mst_optimal_ratio_data = []
mst_avg_time_data = []
for line in lines_mst
    m = match(r"Avg time: ([0-9]*\.?[0-9]*)", line)
    if m != nothing
        value = parse(Float32, m[1])
        push!(mst_avg_time_data, value)
        continue
    end

    m = match(r"Opt ratio: ([0-9]*\.?[0-9]*)", line)
    if m != nothing
        value = parse(Float32, m[1])
        push!(mst_optimal_ratio_data, value)
        continue
    end
end

lines_rand = readlines("./results/tabu/mst_vs_rand/rand.txt")

rand_optimal_ratio_data = []
rand_avg_time_data = []
for line in lines_rand
    m = match(r"Avg time: ([0-9]*\.?[0-9]*)", line)
    if m != nothing
        value = parse(Float32, m[1])
        push!(rand_avg_time_data, value)
        continue
    end

    m = match(r"Opt ratio: ([0-9]*\.?[0-9]*)", line)
    if m != nothing
        value = parse(Float32, m[1])
        push!(rand_optimal_ratio_data, value)
        continue
    end
end

time = scatter(names, [mst_avg_time_data rand_avg_time_data], label = ["MST"  "Rand"], ylabel = "Avg time [s]")
display(time)
savefig(time, "./plots/mst_vs_rand_time.png")

optimal_ratio = scatter(names, [mst_optimal_ratio_data rand_optimal_ratio_data], label = ["MST"  "Rand"], ylabel = "Ratio to optimal cost")
display(optimal_ratio)
savefig(optimal_ratio, "./plots/mst_vs_rand_opt.png")
```
